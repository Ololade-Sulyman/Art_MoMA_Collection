```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# To read the file to dataframe

```python
data = pd.read_csv('Artworks.csv')
```

```python
data.head()
```

# To know the information about the dataset

```python
data.info()
```

# To clean the dataset

```python
# Find Nan

nan_data = data[data.isna().any(axis=1)]


data = data.replace('NaN', pd.NA)

data.columns = data.columns.str.strip()

df = data.dropna(how='all')
df.head()
```

```python
df['BeginDate'] = df['BeginDate'].str.replace('-', '', regex=True)

df['EndDate'] = df['EndDate'].str.replace('-', '', regex=True)

df['Nationality'] = df['Nationality'].str.replace('(', '').str.replace(')', '')

df['ArtistBio'] = df['ArtistBio'].str.replace('(', '').str.replace(')', '')

df['Gender'] = df['Gender'].str.replace('(', '').str.replace(')', '')


```

 **To check the date format**

```python
df['DateAcquired']
```

**To convert the date[DateAcquired] to datetime**

```python
# Convert the date to datetime
df['DateAcquired'] = pd.to_datetime(df['DateAcquired'])
```

```python
# Extract month, year
df['Acquired_Month'] = df['DateAcquired'].dt.month.fillna(0).astype(int)
```

```python
df['Acquired_Year'] = df['DateAcquired'].dt.year.fillna(0).astype(int)



df.head()
```

# Question 2: Types of artworks that are most common

```python
# To get Artwork that most common
Artwork_counts = df['Classification'].value_counts().reset_index()

print(Artwork_counts)
```

```python
# Sort the data and select the top 10 
top_15_Artworks = Artwork_counts.nlargest(15, 'count')

# Reverse the order for the barh plot 
top_15_Artworks = top_15_Artworks.iloc[::-1]

print(top_15_Artworks)
```

```python

# Create a horizontal bar plot
#plt.barh(top_15_Artworks['Classification'], top_15_Artworks['count']) 
sns.barplot(data=top_15_Artworks[::-1], x= top_15_Artworks['count'], y= top_15_Artworks['Classification'], hue='count', palette= 'dark:b_r')
plt.xlabel('Count') 
plt.ylabel('Artwork Type') 
plt.title('Top 15 Most Common Artwork Types')
plt.show()
```

 **The above visualization shows that Photograph is the most common Artwork,
 followed by Print and then  Illustrated Book**

# Question 3:  How modern are the Artworks at the Museum

```python
# To get how modern are the Artworks
import datetime

current_year = datetime.datetime.now().year

# Calculating the Age of each Artwork
df['Age'] = current_year - df['Acquired_Year']

# To classify Artworks as modern, modern threshold = 100
modern_threshold = 100
df['modern'] = df['Age']<= modern_threshold

print(df)
```

```python
# Monthly trend
monthly_trend = df['Acquired_Month'].value_counts()

monthly_trend = monthly_trend.sort_index()
monthly_trend
```
